Here we have loaded outputs for 1 video and 3 clips (0, 10, and 20) from each stage of the model. If you want to see the whole video this is mouse071_task1_annotator1.mp4.

```python
import torch
import numpy as np
from IPython.display import Video, display
import sklearn
```

```python
# load in all predictoins and ground truth data
demo_ifaction_prediction = np.load('demo_ifaction_pred.npy')
demo_embeddings = torch.load("demo_embeddings.pt", map_location=torch.device('cpu'))
demo_clip_predictions = np.load("demo_clip_pred.npy")
demo_action_prediction = np.load("demo_action_pred.npy")

demo_ground_truth_ifactions = np.load('demo_ground_truth_ifactions.npy')
demo_ground_truth_actions = np.load('demo_ground_truth_actions.npy')
```

```python
# view a video clip (sample0, sample10, or sample20)
video_path = 'sample10.mp4'
display(Video(data=video_path))
```

```python
def get_metrics(pred, labels):
    accur = sklearn.metrics.accuracy_score(pred, labels)
    f1_scores = sklearn.metrics.f1_score(pred, labels, average='macro')
    rec_scores = sklearn.metrics.precision_score(pred, labels, average='macro')
    prec_scores = sklearn.metrics.recall_score(pred, labels, average='macro')
    print('Accuracy:', accur, 'F1:', f1_scores, 'Recall:', rec_scores, 'Precision:', prec_scores)
```

```python
print('Stage 1')
get_metrics(demo_ifaction_prediction, demo_ground_truth_ifactions)
print('Stage 2')
get_metrics(demo_action_prediction, demo_ground_truth_actions)
```
